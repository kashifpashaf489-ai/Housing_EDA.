##**USING ENCODER TO CONVERT OBJECT TO NUMERIC **##

```python
import sklearn
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

##**Insight** = LabelEncoder: Converts categorical labels into numerical integers. Typically used for target variables. OneHotEncoder: Converts categorical variables into binary (0 or 1) columns, with each column representing a category. ColumnTransformer: Allows applying different preprocessing techniques to different subsets of columns in your dataset (e.g., scaling for numerical columns, encoding for categorical ones).

```python
le = LabelEncoder()
```

##**Insight** = Giving short cut for label encoder to use again and again. ##

```python
df = le.fit_transform
print(df)
```

##**Insight** = Fit learns the unique categories in the data and establishes the mapping between categories and numbers. transform then converts the original data into the numerical format based on that mapping. fit_transform combines both steps: it learns the mappings and then directly applies them to transform the data in one go.##

```python
X[:1,:]
```

##**Insight** = Checks the 1st line of dataset. ##

```python
ct = ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[-1])],remainder='passthrough')
```

```python
X = ct.fit_transform(X)
```

##**Insight** = The column transformer gets activated which helps in apply different preprocessing steps to specific columns of a dataset. we applied one-hot encoding to a categorical column and kept the other columns unchanged. -2 refers to the second-to-last column in the feature matrix X. The remainder='passthrough' ensures that remaining columns(except -2 column) are unchanged so it can be used later.##

##**Scaling **##

##Scaling refers to process of transforming features (input data) so that they are on a similar scale or within a specific range.##

##**Normalization **##

x = x - min(x)/max(x) - min(x)

This method is useful, to scale your data to a fixed range between 0 to 1.

Putting data into a particular range(0,1) or (-1,1)


```python
from sklearn.preprocessing import MinMaxScaler
```

```python
normal = MinMaxScaler()
```

```python
X = normal.fit_transform(X)
print(X)
```

##**Insight** = MinMaxScaler is used to normalize the features of a dataset by scaling them to a specific range of 0 to 1. The fit_transform method first calculates the minimum and maximum values of each feature (fit), then scales the data according to those values (transform).##

##**Strandadization **##
x = (x - μ) / σ

μ is the mean of the feature

σ is the standard deviation of the feature

The result is a distribution with a mean of 0 and a standard deviation of 1. The data is centered around 0.
